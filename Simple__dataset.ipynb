```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("iris.csv") # to load csv data set using pandas

print(df.shape) # to see the no of columns and rows in dataset
print(df.columns) # to see the columns name
print(df.head(5)) # to see the first 5 rows


df.info() #tell about data type of column,null or not ,kitni entries han
df.describe()#minimum value, maximu,average,standard deviation,quartiles

#scatter plot
plt.scatter(df['petal_length'], df['petal_width'])
plt.xlabel("petal length")
plt.ylabel("sepal length")
plt.title("scatter plot for petal length and petal width")
plt.show()

#histogram
df["sepal_length"].hist(bins=30,edgecolor='black',color='pink')
plt.xlabel("sepal length")
plt.ylabel("frequency")
plt.title("histogram for sepal length and its frequency")
plt.show()

#box plot
#outlier wo values jo normal data sy alag ho ya bohat choti hon ya bohat bari its affect the model accuracy
sns.boxplot(df['sepal_width'])
plt.title("Box Plot of Sepal width")
plt.ylabel("Sepal width")
plt.show()

```
