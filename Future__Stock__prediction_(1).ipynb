```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


ticker_symbol = "AAPL"
period = "1y"

stock_data = yf.download(ticker_symbol,period=period,auto_adjust=True)

stock_data["Target"] = stock_data["Close"].shift(-1)

stock_data=stock_data.dropna()

X = stock_data[["Open", "High", "Low", "Volume"]]
Y = stock_data["Target"]

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,shuffle=False)

from sklearn.linear_model import LinearRegression
model = LinearRegression() #instance create Kiya ha model ka

model.fit(X_train,Y_train) # model train ho chuka ha yahan

Y_pred = model.predict(X_test) #test data pr prediction krwani ha

model.score(X_test,Y_test)

plt.plot(Y_test.values, color="blue")

plt.plot(Y_pred, color="red")

plt.title("Actual vs Predicted Close Price")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()




```
